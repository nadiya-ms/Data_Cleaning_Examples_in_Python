# Data exploration and cleaning

```python
# import libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read excel file
df = pd.read_excel("rollingsales_bronx.xlsx")

# preview data
df.head()
```

This data frame include: 1. unnamed columns, 2. missing values in the first three rows, 3. the actual table in the doesn't start until the 4th row. 


```python
# read the data skipping 3 rows this time
df = pd.read_excel("rollingsales_bronx.xlsx", skiprows = 4, header = [0])

df.head()
```

```python
# preview column names
df.columns
```

```python
#Extract columns
cols = df.columns

new_cols = []
#iterate to fix issues with names
for column in cols:
    #to proper case
    proper_cols = column.title()
    #replace space/hyphen with underscore
    proper_cols_hyphen = proper_cols.replace(" ", "_")
    clean_col = proper_cols_hyphen.replace("-", "_")
    #append to empty list
    new_cols.append(clean_col)
#diplay columns
new_cols
```

```python
#replace existing columns in dataframe with new
df.columns = new_cols
df.head()
```

```python
# data types
df.dtypes
```

## Converting Data Types

All data types appear in order except the **Year_Built** column, which should be categorical.

```python
# fix Year_Built column
df.Year_Built = df.Year_Built.astype("str")
df.dtypes
```

## Handling Missing Values

```python
df.shape
```

```python
df.info()
```

```python
cols = df.columns

#plot a heatmap of missing values with seaborn
plt.figure(figsize = (10,5))
sns.heatmap(df[cols].isnull())
plt.savefig("missing_heatmap.png")
```

```python
# get the percentage of missing values in each column
missing_pct = round(df.isnull().mean() * 100, 1)
print(missing_pct)
```

```python
# Drop columns with more than 70% missing values
drop_cols = missing_pct[missing_pct > 70].index
df_new = df.drop(columns = drop_cols)
df_new.head()
```

```python
# Extract columns with mising values between 1 and 30%
replace_cols = missing_pct[(missing_pct>0) & (missing_pct < 30)].index
df_new[replace_cols ]
```

```python
df_new[replace_cols].describe()
```

```python
#Iterate to replace missing values
for col in replace_cols:
    mean_value = df_new[col].mean()
    df_new.fillna(mean_value, inplace = True)

#preview for missing values
df_new.isnull().sum()
```

```python
df_new.shape
```

## Removing Duplicates

```python
# duplicates
duplicates = df_new[df_new.duplicated(df_new.columns)]
duplicates
```

```python
# Drop duplicates
df_new.drop_duplicates(inplace=True)
```
